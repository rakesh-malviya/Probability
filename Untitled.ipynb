# Discrete Random Variables

Given an experiment: rolling a 6 face die. We can get any of the following outcomes [1,2,3,4,5,6]. The set of all possible outcomes is called <b>Sample space</b>. Sample of the above experiment is [1,2,3,4,5,6]. <b>Random Variable</b> associates a value to the outcome of the experiment. <b>A random variable is a real-valued function of the outcome of the experiment.</b>

Given above experiment we can have following random variables:
1. iseven(x) in {1,0} value of random variable is 1 if face f is [2,4,6] else 0 [1,3,5]
    . i.e. $iseven(x) \in \{1,0\}$
2. value(x) in {1,2,3,4,5,6} same as the outcome of die roll
    . i.e. $value(x) \in \{1,2,3,4,5,6\}$
3. ...


## Probability Mass Fuction (PMF)

$PMF$ of $X$ is denoted as $p_X$. Function $p_X(x)$ is probability of event $\{X=x\}$ i.e. random variable $X$ is equal to $x$, where $x$ is one of the possible values $X$

Let the experiment consist of two independent tosses of a fair coin, and let $X$ be the number of heads obtained. Then the $PMF$ of $X$ is

$p_X(x) =
\left\{
	\begin{array}{ll}
		{1 \over 4}  & \mbox{if } x = 0 \mbox{ or } x = 2 \\
		{1 \over 2} & \mbox{if } x = 1 \\
        {0} & \mbox{otherwise}
	\end{array}
\right.$

i.e. <br>
$p_X(0)=1/4$ when outcome is <b>HH</b><br>
$p_X(1)=1/2$ when outcome is <b>HL</b> or <b>LH</b><br>
$p_X(2)=1/4$ when outcome is <b>LL</b><br>
$p_X(x)=0$ where $x\notin\{0,1,2\}$ ...


Calculation of the PMF of a Random Variable X
For each possible value x of X:
1. Collect all the possible outcomes that give rise to the event {X = x}.
2. Add their probabilities to obtain pX(x).

### Bernoulli Random Variable

Consider the toss of a biased coin, which comes up a head with probability p,
and a tail with probability 1 − p. The Bernoulli random variable takes the two
values 1 and 0, depending on whether the outcome is a head or a tail:

$X=
\left\{
	\begin{array}{ll}
		{1} & \mbox{  if head}\\
        {0} & \mbox{  if tail}
	\end{array}
\right.$

<br><br>
Its $PMF$ is :

$p_X(x)=
\left\{
	\begin{array}{ll}
		{p} & \mbox{  if } x = 1\\
        {1-p} & \mbox{  if } x = 0
	\end{array}
\right.$



```python
"""Below we will we will write code of Bernoulli Random Variable. 
class BernoulliSample simulates probability of outcomes when coin
is tossed large number of times
"""
import numpy as np
from collections import Counter
from matplotlib import pyplot as plt

class BernouliSample:
  def __init__(self,p,samples=1000000):
    randamMax = 100
    
    data = np.random.choice(randamMax,samples)
    mask =(data <randamMax*p)
    data = np.zeros(samples,dtype=np.int)
    data[mask]=1
    self.data = data
    self.counter = Counter(self.data)
    
    self.prob = {} #works as PMF function 
    for x,y in self.counter.items():
      self.prob[x] = y/float(samples)
    
  def PMF(self,x):
    return self.prob.get(x,0)
    
```

### The Binomial Random Variable

A biased coin is tossed n times. At each toss, the coin comes up a head with
probability $p$, and a tail with probability $1−p$, independently of prior tosses. Let
X be the number of heads in the n-toss sequence. We refer to X as a <b>binomial
random variable</b> with parameters $n$ and $p$. The $PMF$ of $X$:

$p_X(k) = P(X=k) = {n \choose k}p^k(1-p)^{n-k} ,  k =0,1,2....,n.$



```python
"""Below we will we will write code of Binomial Random Variable. 
class Bernoulli is based on mathematical formula above 
"""
import numpy as np
from collections import Counter
from matplotlib import pyplot as plt
import operator as op

class Binomial:
  def __init__(self,p,n):
    self.prob = {}
    self.p = p
    self.n = n
    self.p_pow_table = np.zeros(n+1)
    self.p_1_pow_table = np.zeros(n+1)
    
    p_pow = 1.0
    for k in range(n+1):
      self.p_pow_table[k]=p_pow
      p_pow *=p
      
    p_1_pow = 1.0
    p_1 = 1.0-p
    for k in range(n+1):
      self.p_1_pow_table[k]=p_1_pow
      p_1_pow *=p_1
      
    for k in range(n+1):        
        self.prob[k] = self.pmfFunc(k)
  
  def ncr(self,n, r):
    r = min(r, n-r)
    if r == 0: return 1
    numer = reduce(op.mul, xrange(n, n-r, -1))
    denom = reduce(op.mul, xrange(1, r+1))
    return numer//denom

  def pmfFunc(self,k):
    n = self.n
    returnVal = self.ncr(n,k)*self.p_pow_table[k]*self.p_1_pow_table[n-k]
    return returnVal
    
    
  def PMF(self,x):
    return self.prob.get(x,0)

b = Binomial(p=0.5,n=9)
sum(b.prob.values())
```

### The Geometric Random Variable

Suppose that we repeatedly and independently toss a biased coin with probability
of a head $p$, where $0 < p < 1$. The geometric random variable is the number
$X$ of tosses needed for a head to come up for the first time. Its $PMF$ is given by:

$p_X(k) = (1-p)^{k-1}p$,  where $k = 1,2,...,$

```python
"""Below we will we will write code of Geometric Random Variable. 
class Geometric is based on mathematical formula above 
"""

class Geometric:
    def __init__(self,p,n=20):
        self.prob = {}
        curProb = p
        p_1 = 1.0 - p
        for k in range(n):
            self.prob[k+1] = curProb
            curProb *= p_1
    
    def PMF(self,x):
        return self.prob.get(x,0)
    
        
        
```

### Poisson Random Variable

A Poisson random variable takes nonnegative integer values. Its PMF is given by


$p_X(k) = e^{-\lambda}{\lambda^k \over k!}$,     $k = 0, 1, 2, . . . ,$

the Poisson $PMF$ with parameter $\lambda$ is a good approximation
for a binomial $PMF$ with parameters $n$ and $p$, provided $\lambda = np$, $n$ is very large,
and $p$ is very small

```python
"""Below we will we will write code of Poisson Random Variable. 
class Poisson is based on mathematical formula above 
"""

class Poisson:
    def __init__(self,p,n=20):
        self.prob = {}
        self.l = float(n*p)
        self.e_l = pow(2.71828,self.l*(-1.0))
        self.l_k = np.zeros(n+1)
        
        curProb = self.e_l
        for k in range(n+1):
            self.prob[k] = curProb
            curProb *= (self.l/(k+1))
            
    
    def PMF(self,x):
        return self.prob.get(x,0)
    
        
        
```
